```python
import pandas as pd
```

```python
# Load experiment results
bag_of_words_results = r"Models\Baseline_Model\Bag_of_Words\results_bag\experiment_results.csv"
tf_idf_results = r"Models\Baseline_Model\TF-IDF\results_tf-idf\experiment_results.csv"
w2v_results = r"Models\Word_Embedding\results_w2v\experiment_results.csv"
bert_results = r"Models\Advanced_Embeddings\Bert_Fine_tuned\results_bert\experiment_results.csv"
```

```python
df_bow = pd.read_csv(bag_of_words_results)
df_tfidf = pd.read_csv(tf_idf_results)
df_w2v = pd.read_csv(w2v_results)
df_bert = pd.read_csv(bert_results)

# =========================
# Add model category labels
# =========================

df_bow["representation"] = "Bag_of_Words"
df_tfidf["representation"] = "TF_IDF"
df_w2v["representation"] = "Word2Vec"
df_bert["representation"] = "BERT"

# =========================
# Concatenate all results
# =========================

df_all = pd.concat(
    [df_bow, df_tfidf, df_w2v, df_bert],
    ignore_index=True
)

# =========================
# Inspect combined dataframe
# =========================

print(df_all.head())
print("\nShape:", df_all.shape)
print("\nColumns:")
print(df_all.columns)

# =========================
# Optional: Save merged results
# =========================

df_all.to_csv("all_experiment_results.csv", index=False)

print("\nCombined dataframe saved as:")
print("all_experiment_results.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("cleaned_all_results.csv")

# CLEAN COLUMN NAMES
df.columns = [c.strip() for c in df.columns]

# CREATE CLEAN REPRESENTATION LABELS
def create_representation_label(row):

    rep = str(row["Representation"])
    model = str(row["Model"])

    # Classical Representations
    if rep == "BagOfWords":
        return "Bag_of_Words"

    elif rep == "TF-IDF":
        return "TF_IDF"

    elif rep == "Word2Vec":
        return "Word2Vec"

    # Frozen BERT
    elif "BERT_CLS_Frozen" in rep:
        return "BERT_CLS_Frozen"

    elif "BERT_Mean_Frozen" in rep:
        return "BERT_Mean_Frozen"

    # Fine Tuned BERT
    elif "BERT_Finetuned" in rep or "BERT_Fine" in model:
        return "BERT_FineTuned"

    return rep


df["Representation_Label"] = df.apply(
    create_representation_label,
    axis=1
)

df["Config"] = (
    df["Representation_Label"]
    + "\n"
    + df["Model"]
)

# CHECK LABELS
print("\nUnique Representations:")
print(df["Representation_Label"].unique())

print("\nUnique Models:")
print(df["Model"].unique())

plt.rcParams["figure.figsize"] = (20, 9)
plt.rcParams["font.size"] = 10

# 1. F1-SCORE BY CONFIGURATION + PIPELINE
grouped_f1 = (
    df.groupby(
        ["Config", "Pipeline"]
    )["F1-Score"]
    .mean()
    .unstack()
)

ax = grouped_f1.plot(kind="bar")

# Add labels on bars
for container in ax.containers:

    ax.bar_label(
        container,
        fmt="%.3f",
        padding=2,
        fontsize=7
    )

plt.title(
    "F1-Score by Representation, Classifier, and Pipeline",
    fontsize=16
)

plt.ylabel("F1-Score")
plt.xlabel("Representation + Classifier")

plt.xticks(rotation=20)

plt.legend(
    title="Pipeline",
    bbox_to_anchor=(1.02, 1),
    loc="upper left"
)

plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.4
)

plt.tight_layout()
plt.show()

# =====================================================
# 2. ACCURACY BY CONFIGURATION + PIPELINE
# =====================================================

grouped_acc = (
    df.groupby(
        ["Config", "Pipeline"]
    )["Accuracy"]
    .mean()
    .unstack()
)

ax = grouped_acc.plot(kind="bar")

for container in ax.containers:

    ax.bar_label(
        container,
        fmt="%.3f",
        padding=2,
        fontsize=7
    )

plt.title(
    "Accuracy by Representation, Classifier, and Pipeline",
    fontsize=16
)

plt.ylabel("Accuracy")
plt.xlabel("Representation + Classifier")

plt.xticks(rotation=20)

plt.legend(
    title="Pipeline",
    bbox_to_anchor=(1.02, 1),
    loc="upper left"
)

plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.4
)

plt.tight_layout()
plt.show()

# =====================================================
# 3. PIPELINE COMPARISON INSIDE EACH REPRESENTATION
# =====================================================

grouped_pipeline = (
    df.groupby(
        ["Pipeline", "Representation_Label"]
    )["F1-Score"]
    .mean()
    .unstack()
)

ax = grouped_pipeline.plot(kind="bar")

for container in ax.containers:

    ax.bar_label(
        container,
        fmt="%.3f",
        padding=2,
        fontsize=7
    )

plt.title(
    "Pipeline Comparison Across Representations",
    fontsize=16
)

plt.ylabel("Average F1-Score")
plt.xlabel("Pipeline")

plt.xticks(rotation=0)

plt.legend(
    title="Representation",
    bbox_to_anchor=(1.02, 1),
    loc="upper left"
)

plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.4
)

plt.tight_layout()
plt.show()

# =====================================================
# 4. CLASSIFIER COMPARISON
# =====================================================

grouped_classifier = (
    df.groupby(
        ["Representation_Label", "Model"]
    )["F1-Score"]
    .mean()
    .unstack()
)

ax = grouped_classifier.plot(kind="bar")

for container in ax.containers:

    ax.bar_label(
        container,
        fmt="%.3f",
        padding=2,
        fontsize=7
    )

plt.title(
    "Classifier Comparison by Representation",
    fontsize=16
)

plt.ylabel("Average F1-Score")
plt.xlabel("Representation")

plt.xticks(rotation=15)

plt.legend(
    title="Classifier",
    bbox_to_anchor=(1.02, 1),
    loc="upper left"
)

plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.4
)

plt.tight_layout()
plt.show()

# =====================================================
# 5. HEATMAP
# =====================================================

pivot_table = df.pivot_table(
    values="F1-Score",
    index="Config",
    columns="Pipeline",
    aggfunc="mean"
)

fig, ax = plt.subplots(figsize=(14, 10))

heatmap = ax.imshow(
    pivot_table.values,
    aspect="auto"
)

# Axis labels
ax.set_xticks(np.arange(len(pivot_table.columns)))
ax.set_yticks(np.arange(len(pivot_table.index)))

ax.set_xticklabels(pivot_table.columns)
ax.set_yticklabels(pivot_table.index)

# Add values
for i in range(len(pivot_table.index)):
    for j in range(len(pivot_table.columns)):

        ax.text(
            j,
            i,
            f"{pivot_table.values[i, j]:.3f}",
            ha="center",
            va="center",
            fontsize=8
        )

plt.title(
    "F1-Score Heatmap by Configuration and Pipeline",
    fontsize=16
)

cbar = plt.colorbar(heatmap)
cbar.set_label("F1-Score")

plt.tight_layout()
plt.show()

# =====================================================
# 6. ACCURACY VS INFERENCE TIME
# =====================================================

fig, ax = plt.subplots(figsize=(14, 8))

for config in df["Config"].unique():

    subset = df[
        df["Config"] == config
    ]

    ax.scatter(
        subset["Inference_Time_Seconds"],
        subset["Accuracy"],
        label=config,
        s=140
    )

    # Add pipeline labels
    for _, row in subset.iterrows():

        ax.text(
            row["Inference_Time_Seconds"],
            row["Accuracy"],
            row["Pipeline"],
            fontsize=7
        )

plt.xlabel("Inference Time (Seconds)")
plt.ylabel("Accuracy")

plt.title(
    "Accuracy vs Inference Time",
    fontsize=16
)

plt.legend(
    bbox_to_anchor=(1.02, 1),
    loc="upper left"
)

plt.grid(
    linestyle="--",
    alpha=0.4
)

plt.tight_layout()
plt.show()

# =====================================================
# 7. TOP 15 CONFIGURATIONS
# =====================================================

top_15 = (
    df.sort_values(
        by="F1-Score",
        ascending=False
    )
    .head(15)
)

print("\n===================================")
print("TOP 15 CONFIGURATIONS")
print("===================================")

print(
    top_15[
        [
            "Representation_Label",
            "Model",
            "Pipeline",
            "Accuracy",
            "Precision",
            "Recall",
            "F1-Score",
            "Inference_Time_Seconds"
        ]
    ]
)


```
